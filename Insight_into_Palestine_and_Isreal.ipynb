Let's seperate Isreal from palestine from our dataset



```python
# Filter the dataset to select only rows where 'Country' is 'Israel'
Israel = data[data['Country'] == 'Israel']

# Display the first few rows of the Israel dataset
Israel.head(5)
```

```python
# Filter the dataset to select only rows where 'Country' is 'Palestine'
palestine = data[data["Country"] == 'Palestine']

# Display the first few rows of the Palestine dataset
palestine.head()
```

Israel Cleaning and Viz

```python
for i in ['GDP (in USD)', 'Agricultural Output (in USD)', 'IT Output (in USD)']:
    # Apply the conversion only to non-numeric (string) values using .loc
    Israel.loc[Israel[i].apply(lambda x: isinstance(x, str)), i] = Israel[i].apply(lambda x: float(x.split(' ')[0]) * 1e9 if isinstance(x, str) else x)
```

```python
Israel.head()
```

```python
# Set the 'Year' column as the index for better data visualization
Israel = Israel.set_index("Year")

# Display the first few rows of the Israel dataset with the updated index
Israel.head()
```

```python
Israel.columns
```

```python
for i in ['Population','Active Military Personnel','Reserve Military Personnel','Number of Tanks','Number of Armoured Vehicles']:
    Israel[i] = Israel[i].str.replace(',','')
    Israel[i] = Israel[i].astype('float')
```

```python
import matplotlib.pyplot as plt

# Define the number of rows and columns for subplots
r = 3
c = 5
it = 1

# Create a color palette for the lines
color_palette = ['b', 'g', 'r', 'c', 'm', 'y', 'k']

# Iterate over the columns (excluding 'Country')
for i in Israel.drop('Country', axis=1).columns:
    plt.subplot(r, c, it)
    plt.title(i)

    # Use a custom color from the palette
    color = color_palette[it % len(color_palette)]

    # Plot the data using a line plot
    Israel[i].plot(kind='line', color=color, label=i, linewidth=2, marker='o')

    plt.xlabel('Year')
    plt.ylabel(i)
    plt.grid(True)

    # Add a legend
    plt.legend(loc='best')

    # Increment the subplot counter
    it += 1

# Use tight layout for enough spacing
plt.tight_layout()

# Add a main title
plt.suptitle('Economic Trends Over Time (Israel)')

# Adjust the overall figure size
plt.subplots_adjust(top=2.0, right=4)

# Adjust the bottom and top margins
plt.subplots_adjust(bottom=0.50, top=4.0)

plt.savefig('Israel.png')

# Show the plots
plt.show()
```

#Compare Some variables btw Israel and palestine

For maternal mortality rate, let's visualize the difference btw this two countries



```python
import matplotlib.pyplot as plt
#
# both DataFrames i.e palestine and Israel have a 'Year' column for the x-axis.

plt.figure(figsize=(10, 6))

# Select the data for 'Maternal Mortality Rate (per 100,000 live births)' for Israel and Palestine
israel_mmr = Israel['Maternal Mortality Rate (per 100,000 live births)']
palestine_mmr = palestine['Maternal Mortality Rate (per 100,000 live births)']

# Create a bar plot for the Maternal Mortality Rate
plt.bar(Israel.index, israel_mmr, width=0.4, label='Israel', color='blue')
plt.bar(palestine.index + 0.4, palestine_mmr, width=0.4, label='Palestine', color='green')

plt.title('Maternal Mortality Rate Comparison')
plt.xlabel('Year')
plt.ylabel('Maternal Mortality Rate (per 100,000 live births)')
plt.legend()
plt.grid(axis='y')
plt.show()
```
